```python
class LinearRegression:
    def __init__(self):
        self._w = 0
        self._b = 0
        self.lr = 0.01
        self.n = 0
        self.epoch = 1000

    def fit(self, X_train, y_train):
        self.n = len(X_train)

        y_prd = self._w * X_train + self._b

    def loss(self, y_train, y_prd):
        mse = sum((y_train - y_prd) ** 2) * 1 / self.n

    def compute_grad(self, X_train, y_train, y_prd):

        dw = (-2 / self.n) * np.sum(X_train * (y_train - y_prd) ** 2)
        db = (-2 / self.n) * np.sum((y_train - y_prd) ** 2)

        self.update_parm(dw, db)

    def update_parm(self, dw, db):
        w -= self.lr * dw
        b -= self.lr * db

    def pred(self, X_test):
        return self._w * X_test + self._b
```
