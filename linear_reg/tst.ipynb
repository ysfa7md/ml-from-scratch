```python
from linear_regression import LinearRegression
from mult_linear_regression import MulLinearRegression
import numpy
```

```python
X = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

model = LinearRegression(lr=0.01, epochs=1000)
model.fit(X, y)

print(model.predict([6]))
```

```python
X = [[1, 2], [2, 3], [3, 4], [4, 5]]

y = [8, 13, 18, 23]

model = MulLinearRegression(lr=0.01, epochs=1000)
model.fit(X, y)

print(model.predict([[5, 6]]))
```

```python
import seaborn as sns
sns.lineplot(model.losses)
```
