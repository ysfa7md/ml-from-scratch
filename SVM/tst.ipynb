```python
from svm import SVM
import numpy as np

# Example dataset
X = np.array([[2, 3], [1, 1], [2, 1], [5, 5], [6, 6]])
y = np.array([-1, -1, -1, 1, 1])

# Scale features
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Train SVM
model = SVM(c=1.0, lr=0.01, epochs=100)
model.fit(X, y)

# Predict
preds = model.predict(X)
print("Predictions:", preds)
print("Accuracy:", model.accuracy(X, y))
```

```python
from sklearn.datasets import load_iris
import numpy as np
from sklearn.model_selection import train_test_split

# Load Iris
data = load_iris()
X = data.data
y = (data.target == 0).astype(int)  # Setosa vs rest
y = np.where(y == 0, -1, 1)  # map to -1/+1

# Scale features
X = (X - X.mean(axis=0)) / X.std(axis=0)

# split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=43, stratify=y
)

# train
model = SVM(c=1.0,lr=0.1, epochs=500)
model.fit(X_train, y_train)

# test
preds = model.predict(X_test)


print("Accuracy:", model.accuracy(X_test, y_test))
```
