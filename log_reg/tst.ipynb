```python
from logistic_regression import LogisticRegression
import numpy as np

# simple dataset (linearly separable)
X = np.array([[1, 2], [2, 3], [3, 3], [6, 5], [7, 7], [8, 6]])

y = np.array([0, 0, 0, 1, 1, 1])

# optional: scale features
X = (X - X.mean(axis=0)) / X.std(axis=0)

model = LogisticRegression(lr=0.1, epochs=200)
model.fit(X, y)

preds = model.predict(X)

print("Predictions:", preds)
print("Accuracy:", model.accuracy(X,y))
```

```python
from sklearn.datasets import load_iris
import numpy as np
from sklearn.model_selection import train_test_split

# load data
data = load_iris()

X = data.data
y = data.target

# binary classification (setosa vs rest)
y = (y == 0).astype(int)

# scale features
X = (X - X.mean(axis=0)) / X.std(axis=0)

# split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=43,stratify=y
)

# train
model = LogisticRegression(lr=0.1, epochs=500)
model.fit(X_train, y_train)

# test
preds = model.predict(X_test)


print("Accuracy:", model.accuracy(X_test,y_test))
```
